```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 
cohort_df = pd.read_csv('../data/cohort_results.csv')

#    ()
cohort_pivot = cohort_df.pivot(
    index='cohort_month', 
    columns='month_number', 
    values='retention_rate'
)

#  
plt.figure(figsize=(15, 8))
sns.heatmap(cohort_pivot, annot=True, fmt='.1f', cmap='RdYlGn', 
            vmin=0, vmax=100, cbar_kws={'label': 'Retention Rate (%)'})
plt.title('Cohort Analysis: Customer Retention Rate', fontsize=16)
plt.xlabel('Months Since First Purchase', fontsize=12)
plt.ylabel('Cohort Month', fontsize=12)
plt.tight_layout()
plt.savefig('../docs/cohort_heatmap.png', dpi=300)
plt.show()

#  
print("    :")
print(f"- 1   : {cohort_df[cohort_df['month_number']==1]['retention_rate'].mean():.1f}%")
print(f"- 3   : {cohort_df[cohort_df['month_number']==3]['retention_rate'].mean():.1f}%")
print(f"-    : {cohort_pivot.iloc[:, 3].idxmax()}")
```
